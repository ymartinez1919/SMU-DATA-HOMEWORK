```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sqlalchemy import create_engine, inspect
```

```python

path = "sqlite:///Resources/hawaii.sqlite"
engine = create_engine(path)

#establish connection
conn = engine.connect()
```

```python
inspector = inspect(engine)

tables = inspector.get_table_names()
tables
```

```python
# /api/v1.0/precipitation


# Convert the query results to a dictionary using date as the key and prcp as the value.


# Return the JSON representation of your dictionary.
```

```python
query = """
        SELECT
            station,
            date,
            prcp
        FROM
            measurement
       ORDER BY
           station asc,
           date asc;
        """
df = pd.read_sql(query, conn)
df
```

```python
# /api/v1.0/tobs
# Query the dates and temperature observations of the most active station for the last year of data.
# Return a JSON list of temperature observations (TOBS) for the previous year.

#from Station Analysis we know that max date is '2017-08-23' so a year of data is '2016-08-23'
#from Station Analysis we know that station id 'USC00519281' has the highest number of observations,  with 2772 observations
```

```python
query = """
        SELECT
          station,
          date,
          tobs
          
        FROM
            measurement
        WHERE
            date >= '2016-08-23'
        AND
            station = 'USC00519281'
        ORDER BY
            date asc;
        """
df = pd.read_sql(query, conn)
df.head()
```

```python
# /api/v1.0/<start>
# When given the start only, calculate TMIN, TAVG, and TMAX for all dates greater than and equal to the start date.
```

```python
start="2016-08-22"

query = f"""
        SELECT
            min(tobs) as TMIN,
            avg(tobs) as TAVG,
            max(tobs) as TMAX
        FROM
            measurement
        WHERE
            date >= '{start}'
        """
df = pd.read_sql(query, conn)
df
```

```python
# /api/v1.0/<start>/<end>
# When given the start and the end date, calculate the TMIN, TAVG, and TMAX for dates between the start and end date inclusive.
```

```python
start="2016-08-22"
end="2016-12-12"

query = f"""
        SELECT
            min(tobs) as TMIN,
            avg(tobs) as TAVG,
            max(tobs) as TMAX
        FROM
            measurement
        WHERE
            date >= '{start}'
        AND 
            date <= '{end}'
        """
df = pd.read_sql(query, conn)
df
```
