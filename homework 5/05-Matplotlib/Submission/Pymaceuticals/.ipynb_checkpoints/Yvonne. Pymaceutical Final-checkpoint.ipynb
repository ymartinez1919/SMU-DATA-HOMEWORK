```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import seaborn as sns

# linear modelling
import statsmodels.api as sm
from scipy.stats import linregress
from sklearn.linear_model import LinearRegression


# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
# They have the Mouse ID in common , using outer to not loose any dat while merging 
merge_df = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="outer")

# Display the data table for preview
merge_df.head()
```

```python
# Checking the number of mice.
#this is going to give me duplicates
num_mice = len(merge_df['Mouse ID'])
num_mice
```

```python
#Show in summary table
number_of_mice_duplicates = pd.DataFrame({"Number of Mice(Duplicates)" : [num_mice]})

number_of_mice_duplicates
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
#use loc to find the duplicate mouse id
duplicate_mice = merge_df.loc[merge_df.duplicated(subset=["Mouse ID","Timepoint"]),"Mouse ID"].unique()

duplicate_mice
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID. used isin
#https://www.statology.org/pandas-drop-rows-with-value/

nodupe_merge_df = merge_df[merge_df["Mouse ID"].isin(duplicate_mice)==False]

nodupe_merge_df
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.
```

```python

#Create summary table for all drug regimes and their count of lines 
# use the non-dupe data set
summary_stat = pd.DataFrame(nodupe_merge_df.groupby("Drug Regimen").count())

summary_stat
```

```python
summary_stat.columns
```

```python
#create my mean, median, variance, standard deviation, and SEM of the tumor volume with goupby of Drug Regimen
#rounding to make it a little cleaner
summary_stat['Mean'] = nodupe_merge_df.groupby("Drug Regimen").mean()["Tumor Volume (mm3)"].round(2)
summary_stat['Median'] = nodupe_merge_df.groupby("Drug Regimen").median()["Tumor Volume (mm3)"].round(2)
summary_stat['Variance'] = nodupe_merge_df.groupby("Drug Regimen").var()["Tumor Volume (mm3)"].round(2)
summary_stat['Standard Deviation'] = nodupe_merge_df.groupby("Drug Regimen").std()["Tumor Volume (mm3)"].round(2)
summary_stat['SEM'] = nodupe_merge_df.groupby("Drug Regimen").sem()["Tumor Volume (mm3)"].round(2)

# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
summary_statistics = summary_stat[["Mouse ID", "Mean", "Median", "Variance", "Standard Deviation", "SEM"]]

summary_statistics
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line

sum2 = nodupe_merge_df.groupby('Drug Regimen')

#single line of code to produce same graph above
summary_stats_table_2 = sum2.agg(['count','mean','median','var','std','sem'])["Tumor Volume (mm3)"].round(2)
summary_stats_table_2
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
```

```python
# Put treatments into a list for for loop (and later for plot labels)


# Create empty list to fill with tumor vol data (for plotting)


# Calculate the IQR and quantitatively determine if there are any potential outliers. 

    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    
    
    # add subset 
    
    
    # Determine outliers using upper and lower bounds
    
```

```python
#original merged data without dupes
nodupe_merge_df.head()
```

```python
#make each of the requested drugs their own df
capomulin_df = nodupe_merge_df.loc[nodupe_merge_df["Drug Regimen"] == "Capomulin",:]
ramicane_df = nodupe_merge_df.loc[nodupe_merge_df["Drug Regimen"] == "Ramicane",:]
infubinol_df = nodupe_merge_df.loc[nodupe_merge_df["Drug Regimen"] == "Infubinol",:]
ceftamin_df = nodupe_merge_df.loc[nodupe_merge_df["Drug Regimen"] == "Ceftamin", :]

#for testing
#capomulin_df
```

```python
#starting with Capomulin
capomulin_last = capomulin_df.groupby('Mouse ID').max()['Timepoint']
capomulin_lvol = pd.DataFrame(capomulin_last)
capomulin_merge = pd.merge(capomulin_lvol, nodupe_merge_df, on=("Mouse ID","Timepoint"),how="left")
capomulin_merge.head()
```

```python

tumor_ca = capomulin_merge["Tumor Volume (mm3)"]


# find my quartiles 
quartiles = tumor_ca.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

#Determine outliers using upper and lower bounds
lower_bound = lowerq - (1.5* iqr)
upper_bound = upperq + (1.5* iqr)

print(f"Capomulin lower bound is {lower_bound} and upper is {upper_bound}. This means that potential outliers are either below {lower_bound} and above {upper_bound}.")
print(f"Capumulin lower quartile is {lowerq}")
print(f"Capumulin upper quartile is {upperq}")
```

```python
#ramicane next with same process from above
#ramicane_df
ramicane_last = ramicane_df.groupby('Mouse ID').max()['Timepoint']
ramicane_lvol = pd.DataFrame(ramicane_last)
ramicane_merge = pd.merge(ramicane_lvol, nodupe_merge_df, on=("Mouse ID","Timepoint"),how="left")
ramicane_merge.head()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
```

```python
#already made data frame for just Capomulin above for boxplots
#capomulin_df = nodupe_merge_df.loc[nodupe_merge_df["Drug Regimen"] == "Capomulin",:]
#have my average tumor volume from scatter plot
#avg_ca_vol =capomulin_df.groupby(['Mouse ID']).mean()
data_corr = avg_ca_vol
corrs = data_corr.corr()
corrs
```

```python
sns.heatmap(corrs)
plt.title("Correlation Heat map")
plt.show()
```

```python
corr=(st.pearsonr(avg_ca_vol['Weight (g)'],avg_ca_vol['Tumor Volume (mm3)'])[0])
corr
print(f"The correlation between mouse weight and tumor volume with Capomulin is {corr}")
```

```python
#linear regression model like in class activity day 3 act 9

x_values = data_corr['Weight (g)']
y_values = data_corr['Tumor Volume (mm3)']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values,color="#A5243D",marker="o")
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(20,35),fontsize=15,color="#0A0903")
plt.xlabel('Weight (g)',fontsize = 14,fontweight = "bold",fontname="Times New Roman")
plt.ylabel('Average Tumor Volume (mm3)',fontsize = 14,fontweight = "bold",fontname="Times New Roman")
plt.title('Capomulin:Mouse Weight vs Avg Tumor Volume',fontsize = 20, fontweight = "bold",fontname="Times New Roman", color = "#A5243D")
print(f"The r-squared is: {rvalue**2}")
plt.show()
```
