# Gender Demographics

```python
#first need count of each individual Gender
    #df.Gender.value_counts(), This counts duplicates players because they can make multiple purchases
```

```python
#reference original table to make a new 
df.head()
```

```python
#new table with columns Wanted
new_columns = [ "SN", "Age", "Gender"]

unique_players = df.loc[:, new_columns]
unique_players.head()
```

```python
#using link below to drop duplcate values so each SN is counted once
#https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop_duplicates.html
unique_players.drop_duplicates()

#reset index to restart the count of rows
unique_players.drop_duplicates().reset_index(drop=True)

#make my variable 
players = unique_players.drop_duplicates().reset_index(drop=True)

#count players by gender
players.Gender.value_counts()
```

```python
#Percentage and Count of Male Players = (count of gender players)/ total players
#Percentage and Count of Female Players = (count of gender players)/ total players
#Percentage and Count of Other / Non-Disclosed = = (count of gender players)/ total players
players.Gender.value_counts()/ len(players)
```

```python
#make Gender Demographics summary table 

#2 variables. count and percent
gender_count = players.Gender.value_counts()
gender_percent = round(players.Gender.value_counts()/ len(players) * 100, 2)

#combine my 2 series using link below
#https://stackoverflow.com/questions/18062135/combining-two-series-into-a-dataframe-in-pandas/18062521#18062521

gender_summary = pd.concat([gender_count, gender_percent], axis=1)
#chamge my column names to match 
gender_summary.columns=["Total Count", "Percentage of Players"]
gender_summary
```

# Purchasing Analysis (Gender)

```python
#Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
#Create a summary data frame to hold the results
#Optional: give the displayed data cleaner formatting
#Display the summary data frame
```

```python
#working with data set below
df.head()
```

```python
#my column names 
df.columns
```

```python
#easiest to agg to do most calculations
purch_analysis_gender=df.groupby("Gender").agg({"Purchase ID": "count","Price":["mean","sum"]}).round(2)

#fix my columns to match 
purch_analysis_gender.columns = ["Purchase Count","Average Purchase Price", "Total Purchase Value"]
purch_analysis_gender
```

```python
#final calculation avg total purchase per person = total purchase value/ total counts( in summary table prior)
purch_analysis_gender["Avg Total Purchase per Person"]=round(purch_analysis_gender["Total Purchase Value"]/gender_summary["Total Count"],2)
purch_analysis_gender
```

# Age Demographics

```python
#Establish bins for ages
#Categorize the existing players using the age bins. Hint: use pd.cut()
#Calculate the numbers and percentages by age group
#Create a summary data frame to hold the results
#Optional: round the percentage column to two decimal points
#Display Age Demographics Table
```

```python
#Establish bins for ages per starter code
bins = [0,9,14,19,24,29,34,39,120]

#create labels for age bins
group_labels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
```

```python
pd.cut(df.Age, bins, labels=group_labels)
df["Age Group"]=pd.cut(df.Age, bins, labels=group_labels)
df.head()
```

```python
# now just use code used before in purchasing analysis(gender) so we are not duplicating SNs
#new table with columns Want, and include new cloumn created
new_columns = [ "SN", "Age", "Gender","Age Group"]

unique_players = df.loc[:, new_columns]
unique_players.head()
```

```python
#make my variable 
players = unique_players.drop_duplicates().reset_index(drop=True)
```

```python
#2 variables. count and percent for Age group
gender_count = players["Age Group"].value_counts()
gender_percent = players["Age Group"].value_counts()/ len(players)

#summary table
age_demo = pd.concat([gender_count, gender_percent], axis=1)
#chamge my column names to match 
age_demo.columns=["Total Count", "Percentage of Players"]
age_demo
```

```python
#2 variables. count and percent for Age group. use groupby to sort correctly 
gender_count = players.groupby("Age Group").size()
gender_percent = round((players.groupby("Age Group").size()/ len(players)) * 100 ,2)

#summary table
age_demo = pd.concat([gender_count, gender_percent], axis=1)

#change my column names to match 
age_demo.columns=["Total Count", "Percentage of Players"]
age_demo
```

# Purchasing Analysis (Age)

```python
#Bin the purchase_data data frame by age
#Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
#Create a summary data frame to hold the results
#Optional: give the displayed data cleaner formatting
#Display the summary data frame
```

```python
#taking Purchasing analysis(gender) and switching groupby
#easiest to agg to do most calculations
purch_analysis_age=df.groupby("Age Group").agg({"Purchase ID": "count", "Price":["mean","sum"]}).round(2)

#fix my columns to match 
purch_analysis_age.columns = ["Purchase Count","Average Purchase Price", "Total Purchase Value"]
purch_analysis_age
```

```python
#final calculation avg total purchase per person = total purchase value/ total counts( in summary table prior)
purch_analysis_age["Avg Total Purchase per Person"]=round(purch_analysis_age["Total Purchase Value"]/age_demo["Total Count"],2)

#show summary table
purch_analysis_age
```

# Top Spenders

```python
#Run basic calculations to obtain the results in the table below
#Create a summary data frame to hold the results
#Sort the total purchase value column in descending order
#Optional: give the displayed data cleaner formatting
#Display a preview of the summary data frame
```

```python
#taking Purchasing anlalysis(gender) and switching groupby
#easiest to agg to do most calculations
top_spenders=df.groupby("SN").agg({"Purchase ID": "count", "Price":["mean","sum"]}).round(2)

#fix my columns to match 
top_spenders.columns = ["Purchase Count","Average Purchase Price", "Total Purchase Value"]

#need to sort from total purchase value
#https://www.askpython.com/python-modules/pandas/inplace-true-parameter
top_spenders.sort_values(by="Total Purchase Value", ascending=False, inplace=True )

#summary table with just the top 5 lines
top_spenders.head()
```

# Most Popular Items

```python
#Retrieve the Item ID, Item Name, and Item Price columns
#Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value
#Create a summary data frame to hold the results
#Sort the purchase count column in descending order
#Optional: give the displayed data cleaner formatting
#Display a preview of the summary data frame
```

```python
#taking Purchasing analysis(gender) and switching groupby
#easiest to agg to do most calculations
pop_item=df.groupby(["Item ID","Item Name"]).agg({"Purchase ID": "count", "Price":["mean","sum"]}).round(2)
pop_item.columns = ["Purchase Count","Item Price", "Total Purchase Value"]
#need to sort purchase count
#https://www.askpython.com/python-modules/pandas/inplace-true-parameter
pop_item.sort_values(by="Purchase Count", ascending=False, inplace=True )

#summary table with just the top 5 lines
pop_item.head()
```
